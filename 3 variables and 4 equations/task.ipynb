```python
# формируем список вещественных чисел от -10 до 10 с шагом 0.1
a = [i/10 for i in range(-100, 101)]
```

```python
# перебираем числа из списка, выбирая те, которые удовлетворяют условиям уравнений
for i in a:
    for j in a:
        if i + j == 8:
            for k in a:
                if i + k == 13 and k - i == 6:
                    x = i
                    y = j
                    z = k
print('x = ', x, ', y = ', y, ', z = ', z)
```
